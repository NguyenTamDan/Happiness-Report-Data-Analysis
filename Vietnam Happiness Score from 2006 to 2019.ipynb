## Introduction

In this notebook, I am only trying to visualize and compare Ladder Score - also means as Happiness Score - and its score components on Vietnam in Asia and its surrounding neighbors. And I will not analyze the causes of the score in detail.

```python
# Import library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Styling the outputs
from matplotlib import style

font={'family':'sans-serif',
      'weight':'bold',
      'size':20}
plt.rc('font',**font)
style.use('fivethirtyeight')
```

```python
# Import the dataset
world_2021 = pd.read_csv('world-happiness-report-2021.csv')
world_happiness = pd.read_csv('world-happiness-report.csv')
```

```python
world_happiness
```

## Vietnam from 2006 to 2021

```python
Vietnam = world_happiness[world_happiness['Country name'] == 'Vietnam']
```

```python
Vietnam
```

```python
plt.figure(figsize = (20, 10))
sns.lineplot(data = Vietnam, x = 'year', y = 'Life Ladder', marker = 'o')
plt.title('Life Ladder')
plt.show()
```

There are certain peaks with regards to Vietnam's ladder score as seen in the years 2008, 2011, 2012 and 2019. There are also certain drops as seen in the years 2009, 2010, and an overall low in 2013. A increase can be seen from 2013 to 2019. The overall trend of the ladder score in this period fluctuates continuously over the years.

```python
plt.figure(figsize = (20, 10))
sns.lineplot(data = Vietnam, x = 'year', y = 'Log GDP per capita', marker = 'o')
plt.title('Log GDP per capita')
plt.show()
```

Vietnam's GDP is on an increasing trend over the years.

```python
plt.figure(figsize = (20, 10))
sns.lineplot(data = Vietnam, x = 'year', y = 'Social support', marker = 'o')
plt.title('Social Support')
plt.show()
```

There are some distinct lows in 2010, 2013 and 2018. 2013 had the lowest index and some distinct highs in 2006, 2011 and 2016.

```python
plt.figure(figsize = (20, 10))
sns.lineplot(data = Vietnam, x = 'year', y = 'Healthy life expectancy at birth', marker = 'o')
plt.title('Healthy life expectancy at birth')
plt.show()
```

Vietnam's Healthy life expectancy at birth is on an increasing trend over the years.

```python
plt.figure(figsize = (20, 10))
sns.lineplot(data = Vietnam, x = 'year', y = 'Freedom to make life choices', marker = 'o')
plt.title('Freedom to make life choices')
plt.show()
```

There's a slight increasing trend, but there are distincts drops as seen in years 2009 and 2011. The lowest index in this period is in year 2011.There are certain peaks with regards to India's Freedom to make life choices as seen in the years 2007, 2013 and 2019.

```python
plt.figure(figsize = (20, 10))
sns.lineplot(data = Vietnam, x = 'year', y = 'Generosity', marker = 'o')
plt.title('Generosity')
plt.show()
```

Generosity amongst Vietnamese people have increased over time.

```python
plt.figure(figsize = (20, 10))
sns.lineplot(data = Vietnam, x = 'year', y = 'Perceptions of corruption', marker = 'o')
plt.title('Perceptions of Corruption')
plt.show()
```

The years 2007, 2010 and 2011 had distinct drops in this period. There is a decreasing trend over the years.

```python
plt.figure(figsize = (20, 10))
sns.lineplot(data = Vietnam, x = 'year', y = 'Positive affect', marker = 'o')
plt.title('Positive affect')
plt.show()
```

This variable fluctuated countinuosly over the years. The highest index is in the year 2019.

## Vietnam in Asia

```python
world_2021[world_2021['Country name'] == 'Vietnam']
```

```python
southeast_asia = world_2021[world_2021['Regional indicator'] == 'Southeast Asia']
south_asia = world_2021[world_2021['Regional indicator'] == 'South Asia']
east_asia = world_2021[world_2021['Regional indicator'] == 'East Asia']
```

```python
# Add east_asia to south_asia
south_east_asia = south_asia.append(east_asia)
```

```python
# Create asia table
asia = south_east_asia.append(southeast_asia).sort_values(by = 'Ladder score', ascending = False).reset_index()
```

```python
asia
```

* Vietnam is the 9th most happiness country in Asia.
* In the world context, Vietnam ranks 78th.

```python
plt.figure(figsize = (13, 6))
plt.xticks(rotation = 45, ha = 'right')

sns.lineplot(x = 'Country name', y = 'Ladder score', data = east_asia, marker = 'o', label = 'East Asia')
sns.lineplot(x = 'Country name', y = 'Ladder score', data = southeast_asia, marker = 'o', label = 'Southeast Asia')
sns.lineplot(x = 'Country name', y = 'Ladder score', data = south_asia, marker = 'o', label = 'South Asia')


plt.show()
```
